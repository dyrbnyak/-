#1 Завод производитель подшипников заявляет, что изготовленные на станках металлические элементы для подшипников, имеют средний диаметр 10 мм. Используя односторонний критерий с уровнем значимости α=0,05, проверить эту гипотезу. При проверке гипотезы необходимо учесть, что была произведена выборка из n=16 шариков, где среднее значение диаметра равно 10,3 мм, а дисперсия известна и равна 1 мм.

```python
#1

from scipy import stats

#h0 = 10
#h1 > 10

# Заданные значения
mu_0 = 10  # предполагаемый средний диаметр
sigma = 1  # известная дисперсия
n = 16  # размер выборки
alpha = 0.05  # уровень значимости

# поиск критического значения
critical_value = stats.norm.ppf(1 - alpha)

# Рассчитываем границу для принятия или отвержения гипотезы
nabl = mu_0 + critical_value * (sigma / (n ** 0.5))

# Выводим результат
if critical_value < nabl:
    print('Нулевая гипотеза подвержается')
else:
    print('Нулевая гипотеза отвергается')
```

#2 Производитель конфет заявляет, что средний вес коробки конфет составляет 100 г. Из партии извлечена выборка из n=10 коробок и взвешена. Вес каждой коробки соответствует таблице вариантов. Не противоречит ли это утверждению продавца? Используя уровень значимости α=0,001. Вес коробок конфет распределен нормально.

```python
#2

#h0=100
#h1!=100
from scipy import stats
import numpy as np

viborka = np.array([98, 104, 97, 97, 101, 100, 99, 101, 99, 98])
x = 100  
alpha = 0.001  


n = len(viborka)
sample_mean = np.mean(viborka)#среднее
sample_std = np.std(viborka, ddof=1)#дисперсия


t_statistic = (sample_mean - x) / (sample_std / np.sqrt(n))

critical_value = stats.t.ppf(1 - alpha / 2, df = n - 1)

# Выводим результаты
print(f"t-статистика: {t_statistic:.4f}")
print(f"Критическое значение c правой стороны: {critical_value:.4f}")
print(f"Критическое значение c левой стороны: {-critical_value:.4f}")

# Принимаем решение
if t_statistic < critical_value:
    print("Принимуем нулевую гипотезу")
else:
    print("Принимем единичную гипотезу")

```

#3 Произведены n=7 независимых измерений, в результате которых найдено, что ¯x=82,48 мм, а S=0,08. Предположив, что ошибки измерения имеют нормальное распределение проверить с использованием уровня значимости α=0,05 гипотезу H0:σ2=0,01 мм2. против конкурирующей гипотезы. H0:σ2=0,005. В ответе записать разность между фактическим и табличным значениями выборочной характеристики.

```python
#3

from scipy import stats


n = 7
x_bar = 82.48
S = 0.08
alpha = 0.05
sigma0_sq1 = 0.01
sigma0_sq2 = 0.005


chi2_statistic_sq1 = (n - 1) * (S ** 2) / sigma0_sq1
chi2_statistic_sq2 = (n - 1) * (S ** 2) / sigma0_sq2

chi2_critical_right = stats.chi2.ppf(1 - alpha/2, df=n - 1)
chi2_critical_left = stats.chi2.ppf(alpha/2, df=n - 1)

print(f"Наблюдаемое(фактическое) значение статистики для 0.01: {chi2_statistic_sq1:.4f}")
print(f"Наблюдаемое(фактическое) значение статистики 0.005: {chi2_statistic_sq2:.4f}")
print(f"Правое критическое значение: {chi2_critical_right:.4f}")
print(f"Левое критическое значение: {chi2_critical_left:.4f}")

if chi2_statistic_sq1 < chi2_critical_right or chi2_statistic_sq1 > chi2_critical_left:
    print("Не отвергаем H0: нет оснований считать, что дисперсия отличается от 0.01 мм^2.")
else:
    print("Отвергаем H0: дисперсия не равна 0.01 мм^2.")

```

#4 Стратегия финансовой организации «Не обманешь!» не инвестирует в ценные бумаги если дисперсия годовой доходности более чем 0,04. Произведена выборка из n=52 наблюдений по активу А показала, что выборочная дисперсия ее доходности равна 0,045. Узнать, допустимы ли для данной финансовой организации инвестиционные вложения в актив А на уровне значимости: а) 0,05; б) 0,01.

```python
#4

from scipy import stats


n = 52
S = 0.045
alpha_1 = 0.05
alpha_2 = 0.01
h0 = 0.04

chi2_statistic = (n - 1) * (S ** 2) / h0

chi2_critical = stats.chi2.ppf(1 - alpha_1/2, df=n - 1)

print(f"Критическое значение: {chi2_critical:.4f}")
print(f"Наблюдаемое значение: {chi2_statistic:.4f}")


if chi2_statistic > chi2_critical:
    print("\nНе инвестируем, дисперсия больше 0.04 при значимости 0.05\n")
else:
    print("\nИнвестируем, дисперсия не больше 0.04 при значимости 0.05\n")


chi2_critical = stats.chi2.ppf(1 - alpha_2/2, df=n - 1)

print(f"Критическое значение: {chi2_critical:.4f}")
print(f"Наблюдаемое значение: {chi2_statistic:.4f}")

if chi2_statistic > chi2_critical:
    print("\nНе инвестируем, дисперсия больше 0.04 при значимости 0.01")
else:
    print("\nИнвестируем, дисперсия не больше 0.04 при значимости 0.01")
```

#5 Фирма «Спам» рассылает рекламные буклеты возможным заказчикам. Как показал опыт, вероятность того, что организация получившая буклет, закажет рекламируемое изделие, равна 0,08. Фирма разослала 1000 буклетов новой, улучшенной, формы и получила 100 заказов. На уровне значимости 0,05 выяснить, можно ли считать, что новая форма рекламы существенно лучше прежней.

```python
#5
#h0 > 0.08, новая реклама значительно лучше
#h1 <= 0.08, новая рекалама не лучше старой
import numpy as np
from scipy import stats


p0 = 0.08  
n = 1000  
x = 100 
alpha = 0.05

p_hat = x / n

z_critical = stats.norm.ppf(1 - alpha)

z_nabl = (p_hat-p0)/np.sqrt(0.08*0.92)*np.sqrt(1000)

print(f"Критическое значение: {z_critical:.4f}")
print(f"Наблюдаемое значение: {z_nabl:.4f}")

if z_nabl>z_critical:
    print("Новая форма рекламы существенно лучше прежней.")
else:
    print("Новая форма рекламы не лучше прежней.")
```
